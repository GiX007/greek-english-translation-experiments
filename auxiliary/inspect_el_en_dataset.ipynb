Quick inspection of the OPUS Books el-en dataset:
load data, display a few samples, and check sentence lengths before training.

The dataset is available on Hugging Face: https://huggingface.co/datasets/Helsinki-NLP/opus_books.

```python
# !pip install datasets
```

```python
from datasets import load_dataset

ds = load_dataset(
    "Helsinki-NLP/opus_books",
    "el-en", # language pair
    split="train"
)
```

```python
ex = ds[1]["translation"]
print("EL:", ex["el"])
print("EN:", ex["en"])
```

```python
for i in range(10):
  print("EL:", ds[i]["translation"]["el"])
  print("EN:", ds[i]["translation"]["en"])
  print("-"*40)
```

```python
import pandas as pd

pd.DataFrame({
    "el_len": [len(x["translation"]["el"].split()) for x in ds],
    "en_len": [len(x["translation"]["en"].split()) for x in ds],
}).describe()
```
